## Analysis: Mean Order Number by Department (Subset vs. Full Dataset)

- The **subset (1 million rows)** provides a rough estimate of the mean order_number for each department, but it's based on a partial view of the data.
- The **full dataset** includes all available rows and gives a more accurate, complete view.
- For some departments, the differences between subset and full dataset are small, which suggests that the subset is reasonably representative.
- This highlights the importance of using the full dataset when performing final analyses, especially for departments that aren't well represented in the subset.
  


```python
## Q4
ords_prods_merge['avg_order'] = ords_prods_merge.groupby('user_id')['order_number'].transform('mean')

ords_prods_merge.loc[ords_prods_merge['avg_order'] > 20, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['avg_order'] <= 20) & (ords_prods_merge['avg_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['avg_order'] <= 10, 'loyalty_flag'] = 'New customer'

ords_prods_merge['loyalty_flag'].value_counts()
```

## Price Analysis by Customer Loyalty

I analyzed whether loyal customers buy more expensive products compared to new or regular customers.

### Key Observations:
- **Median product price** is exactly the same (7.4) across all customer groups, indicating that the typical product purchased is consistent.
- **New customers** have the highest mean price (~13.31), suggesting they may experiment with more expensive items during early usage.
- **Regular customers** follow with a mean of aroud 12.03, while loyal customers have the lowest mean (~10.47), possibly reflecting a tendency to repurchase lower-priced or familiar products.
- This pattern may indicate that loyalty leads to more consistent and cost-conscious purchasing, whereas new users explore more premium options.
- The marketing team could use this insight to offer higher-priced product recommendations or promotions to new users, while incentivizing bulk deals or discounts for loyal customers.


```python
## Q6
ords_prods_merge['avg_price_per_user'] = ords_prods_merge.groupby('user_id')['prices'].transform('mean')
ords_prods_merge.loc[ords_prods_merge['avg_price_per_user'] < 10, 'spending_flag'] = 'Low spender'
ords_prods_merge.loc[ords_prods_merge['avg_price_per_user'] >= 10, 'spending_flag'] = 'High spender'
ords_prods_merge['spending_flag'].value_counts()

```

```python
## Q7
ords_prods_merge['median_days'] = ords_prods_merge.groupby('user_id')['days_since_prior_order'].transform('median')

ords_prods_merge.loc[ords_prods_merge['median_days'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[
    (ords_prods_merge['median_days'] > 10) & (ords_prods_merge['median_days'] <= 20),
    'order_frequency_flag'
] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_days'] <= 10, 'order_frequency_flag'] = 'Frequent customer'

ords_prods_merge['order_frequency_flag'].value_counts()
```

## Creating Order Frequency Flags

I analyzed customer behavior based on the median number of days between their orders. Users were labeled as:

- **Frequent customer**: median ≤ 10 days
- **Regular customer**: 10 < median ≤ 20 days
- **Non-frequent customer**: median > 20 days

This helps the Instacart team determine how regularly each customer shops and can be used to send personalized notifications for reordering.
