```python
import pandas as pd
import numpy as np
import os

# Define the project folder path
project_folder = r"/Users/cem/Desktop/Data Immersion/Achievement 4_Project"

# Define the data folder path
data_folder = os.path.join(project_folder, "Data", "4.3_orders_products")

# Load orders.csv
orders_path = os.path.join(data_folder, "orders.csv")
df_ords = pd.read_csv(orders_path, index_col=False)

# Display the first 5 rows
df_ords.head()
```

```python
# Load products.csv
products_path = os.path.join(data_folder, "products.csv")
df_prods = pd.read_csv(products_path, index_col=False)

# Display the first 5 rows
df_prods.head()
```

```python
# Load departments.csv
departments_path = os.path.join(data_folder, "departments.csv")
df_dep = pd.read_csv(departments_path, index_col=False)

# Display the first 5 rows
df_dep.head()
```

```python
# Q2
print(df_ords.columns)
```

```python
# Q2: Change order_number to a string
df_ords['order_number'] = df_ords['order_number'].astype(str)

# Check data types
print(df_ords.dtypes)
```

```python
# Q3: Rename 'order_dow' to 'order_day_of_week' without overwriting the original DataFrame
df_ords_renamed = df_ords.rename(columns={'order_dow': 'order_day_of_week'})

# Display the first 5 rows of the renamed DataFrame
df_ords_renamed.head()
```

```python
# Q4: Calculate frequency of orders by hour
order_hour_frequency = df_ords['order_hour_of_day'].value_counts().sort_index()

# Find the busiest hour
busiest_hour = order_hour_frequency.idxmax()
order_count = order_hour_frequency.max()

# Display the results
print("Frequency of orders by hour:")
print(order_hour_frequency)
print(f"\nThe busiest hour for placing orders is {busiest_hour} with {order_count} orders.")
```

```python
# Q5
# Transpose
df_dep_t = df_dep.T

# Use first row as header
new_header = df_dep_t.iloc[0]

# Create new DataFrame from row 1 onward
df_dep_t_new = df_dep_t[1:]

# Set new header row as columns
df_dep_t_new.columns = new_header

# Display transformed departments DataFrame
df_dep_t_new.head()

# The header is now fixed and the rows-columns reversed with transpose:
```
